```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver

driver = webdriver.Chrome(chromedriver_binary.chromedriver_filename)

driver.get('https://www.youtube.com/@3blue1brown/videos')
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
videos = soup.find_all('ytd-rich-item-renderer')
```

```python
data = []

for video in videos:
    try:
        video_link = 'https://www.youtube.com/' + video.find('a').get('href')
    except:
        video_link = np.nan
    try:
        thumbnail_link = video.find('img').get('src').split('?')[0]
    except:
        thumbnail_link = np.nan
    try:
        duration = video.find('span').get('aria-label')
    except:
        duration = np.nan
    
    data.append([video_link, thumbnail_link, duration])
```

```python
df =  pd.DataFrame(data, columns = ['video_link' , 'thumbnail_link', 'duration'])

df.isnull().sum()
```
